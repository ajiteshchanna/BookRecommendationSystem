```python
ratings = pd.read_csv("BX-Book-Ratings.csv", sep=";", encoding='latin-1', on_bad_lines='skip')
```

```python
ratings.head()
```

```python
ratings.rename(columns=
    {
        "User-ID":"user_id",
        "ISBN":"isbn",
        "Book-Rating":"rating"

    }, inplace=True
)
```

```python
# comparing the size of all the three datasets that we have
print(books.shape,users.shape,ratings.shape)
```

```python
# lets store users who have rated more than 200 books
x = ratings['user_id'].value_counts()>200
```

```python
x
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
books.head()
```

```python
ratings_with_books = ratings.merge(books,on='isbn')
```

```python
number_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
# yeh pta karo ki yha kya ho rha hai
```

```python
number_ratings.rename(columns={"rating":"num_of_rating"},inplace=True)
```

```python
number_ratings.head()
```

```python
final_rating = ratings_with_books.merge(number_ratings, on='title')
```

```python
final_rating.head()
```

```python
# let's take those books which have atleast 50 ratings of user
final_rating = final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.isnull().sum() # therefore no missing values
```

```python
# lets create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot.head()
```

```python
# now, here NaN means that person has given 0 rating, so for that we have to fill it with 0
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
# so we have made a sparse matrix
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
np.where(book_pivot.index == '4 Blondes')[0][0]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
# final_rating['title].value_counts
ids = np.where(final_rating['title'] == 'Harry Potter and the Chamber of Secrets (Book 2)')[0][0]
```

```python
book_names = []
for book_id in suggestion:
    book_names.append(book_pivot.index[book_id])
```

```python
ids_index = []
for name in book_names[0]: 
    ids = np.where(final_rating['title'] == name)[0][0]
    ids_index.append(ids)
```
